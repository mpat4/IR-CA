## Tool to identify stations within CalWQA LOEs

```python
import pandas as pd
import numpy as np
import qgrid
import ipywidgets
```

```python
def get_stations(loes_df, stationlist):
    """function to extract the sample code from each LOE's spatial representation field"""
    loes_df['inCEDEN'] = 0
    loes_df['CEDEN_Stations'] = ""
    for i in stationlist:
        print(i)
        search = '\\b' + i + '\\b'
        loes_df['inCEDEN'] = np.where(loes_df['Spatial Representation'].astype(str).str.contains(search), 1, loes_df['inCEDEN'])
        loes_df['CEDEN_Stations'] = np.where(loes_df['Spatial Representation'].astype(str).str.contains(search), loes_df['CEDEN_Stations'].apply(lambda x: str(i) if x == '' else str(x) + ", " + str(i)), loes_df['CEDEN_Stations'].astype(str))
        
```

```python
loes_df = pd.read_excel("H:/220405_CompReport_DeltaSplits.xlsx")
print(loes_df['Spatial Representation'][0])
loes_df.head(1)
```

```python
station_df = pd.read_excel("C:/data_2024/HelpfulLookups/StationLookUp_CEDEN.xlsx")
#stations that need to be removed because return false results
print("{} number of stations".format(len(station_df.index)))
station_df.head(1)

```

```python
remove = list(range(0,100))
remove = [str(x) for x in remove]

remove2 = ["Veterans", "2002", "44", "50" "1","0", "13", "3", "33", "CA", "4", "5", "SSB", "Park", "Barranca", '80', "39", "2003", "2006", "C", "Solano", "Main", "Discovery", "2001", "Freeport"]
remove.extend(remove2)
station_df = station_df[~station_df['StationCode'].isin(remove)]
station_df = station_df.dropna(subset=['StationCode'])
print("{} number of stations".format(len(station_df.index)))
```

```python
station_list = station_df['StationCode'].tolist()
```

```python
get_stations(loes_df, station_list)
```
